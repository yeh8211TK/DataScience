```python
import requests
from bs4 import BeautifulSoup
import openpyxl

workbook = openpyxl.Workbook()
sheet = workbook.worksheets[0]
book_list = []

# 單一種類所有頁面的書籍
def showbook(url, kind):
    html = requests.get(url).text
    sp = BeautifulSoup(html, "html.parser")
    try:
        pages = int(sp.select(".cnt_page span")[0].text)
        print(kind, "共有", pages, "頁")
        for page in range(1, 1 + pages):
            pageurl = url + "&page=" + str(page)
            print("第", page, "頁:", pageurl)
            showpage(pageurl, kind)
    except:
        print(kind, "共有 1 頁")
        print("第 1 頁:", url)
        showpage(url, kind)
    
# 單一種類每一個頁面的書籍
def showpage(url, kind):
    html = requests.get(url).text
    sp = BeautifulSoup(html, "html.parser")
    res = sp.find("div", {"class": "mod type02_m012 clearfix"})
    items = res.select(".item")
    
    for item in items:
        msg = item.select(".msg")[0]
        title = msg.select("a")[0].text
        author = msg.select("a")[1].text
        publish = msg.select("a")[2].text
        date = msg.find("span").text.split("：")[-1]
        onsale = item.select(".price .set2")[0].text
        
        print("書名:" + title)
        print("作者:" + author)
        print("出版社:" + publish)
        print("出版日期:" + date)
        print("優惠價:" + onsale)
        
        listdata = [kind, title, author, publish, date, onsale]
        book_list.append(listdata)
        
def twobyte(categ):
    if categ < 10:
        categstr = "0" + str(categ)
    else:
        categstr = str(categ)
    return categstr

# 書籍種類的數目
homeurl = "https://www.books.com.tw/web/books_nbtopm_01/?o=1&v=1"
html = requests.get(homeurl).text
sp = BeautifulSoup(html, "html.parser")
res = sp.find("div", {"class": "mod_b type02_l001-1 clearfix"})
hrefs = res.find_all("a")

url = "https://www.books.com.tw/web/books_nbtopm_"
mode = "/?0=1&v=1"
categ = 1

# 顯示一個種類的書籍
for href in hrefs[5:6]:
    # 各個書籍種類的網址
    categ_url = url + twobyte(categ) + mode
    categ_content = href.text
    showbook(categ_url, categ_content)
    categ +=1
    
list_title = ["分類", "書名", "作者", "出版社", "出版日期", "優惠價"]
sheet.append(list_title)

for item_n in book_list:
    sheet.append(item_n)
    
workbook.save("book_all.xlsx")
```

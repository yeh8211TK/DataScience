### 動差法(method of moment)

#### 1. k 階動差的定義

k 階中心動差:

$$\mu_{k}=E\left [\left(X-\mu \right)^{k} \right]=\left\{\begin{matrix}
\sum_{i=1}^{n}\left (x_{i}-\mu \right)^{k}\cdot P\left (x_{i} \right )\\ 
\int_{-\infty}^{\infty}\left(x-\mu \right)^{k}\cdot f\left (x \right )dx
\end{matrix}\right.$$

k 階原點動差:

$$\mu_{k}^{'}=E\left[X^{k}\right]=\left\{\begin{matrix}
\sum_{i=1}^{n}x_{i}^{k}\cdot P\left(x_{i}\right)\\ 
\int_{-\infty}^{\infty}x^{k}\cdot f\left(x\right)dx
\end{matrix}\right.$$

(1) 1 階原點動差:(k=1)

$$\mu_{1}^{'}=E\left[X\right]=\left\{\begin{matrix}
\sum_{i=1}^{n}x_{i}\cdot P\left(x_{i}\right)\\ 
\int_{-\infty}^{\infty}x\cdot f\left(x\right)dx
\end{matrix}\right.=\mu$$

意義: 期望值

期望值的**線性性質**:

- $E\left(aX\right)=a\cdot E\left(X\right)$

- $E\left(aX+bY\right)=a\cdot E\left(X\right)+b\cdot E\left(Y\right)$

- $E\left(a+bX\right)=a+b\cdot E\left(X\right)$

(2) 2 階中心動差:(k=2)

$$\mu_{2}=E\left[\left(X-\mu\right)^{2}\right]=\left\{\begin{matrix}
\sum_{i=1}^{n}\left(x_{i}-\mu\right)^{2}\cdot P\left(x_{i}\right)\\ 
\int_{-\infty}^{\infty}\left(x-\mu\right)^{2}\cdot f\left(x\right)dx
\end{matrix}\right.=Var\left(x\right)=\sigma^{2}$$

意義: 變異數 (= 2 階原點動差 - 1 階原點動差的平方 $= \mu_{2}^{'}-\mu^{2}$)

變異數的性質:

- $Var\left(aX\right)=a^{2}\cdot Var\left(X\right)$

- $Var\left(a+bX\right)=b^{2}\cdot Var\left(X\right)$

- $Var\left(X\pm Y\right)=Var\left(X\right)\pm Var\left(Y\right)$

(3) 3 階中心動差:(k=3)

$$\mu_{3}=E\left[\left(X-\mu\right)^{3}\right]=\left\{\begin{matrix}
\sum_{i=1}^{n}\left(x_{i}-\mu\right)^{3}\cdot P\left(x_{i}\right)\\ 
\int_{-\infty}^{\infty}\left(x-\mu\right)^{3}\cdot f\left(x\right)dx
\end{matrix}\right.$$

意義: 偏態(skewness)

$$\text{偏態係數 = }\frac{\mu_{3}}{\sigma^{3}}$$

(4) 4 階中心動差:(k=4)

$$\mu_{4}=E\left[\left(X-\mu\right)^{4}\right]=\left\{\begin{matrix}
\sum_{i=1}^{n}\left(x_{i}-\mu\right)^{4}\cdot P\left(x_{i}\right)\\ 
\int_{-\infty}^{\infty}\left(x-\mu\right)^{4}\cdot f\left(x\right)dx
\end{matrix}\right.$$

意義: 峰度(kurtosis)

$$\text{峰度係數 = }\frac{\mu_{4}}{\sigma^{4}}=\left\{\begin{matrix}
>3\ (高峽峰)\\ 
=3\ (常態峰)\\ 
<3\ (低闊峰)
\end{matrix}\right.$$

#### 2. 動差生成函數(moment-generating function): $M_{x}\left ( t \right )=E\left ( e^{tx} \right )$

(1) 離散型:

$$M_{x}\left ( t \right )=\sum_{i=1}^{n}e^{tx_{i}}P\left ( X=x_{i} \right )$$

r 階原點動差計算:

$$\begin{equation} 
\begin{aligned}
&\frac{\mathrm{d}^{r} M_{x}\left ( t \right )}{\mathrm{d} t^{r}}=\sum_{i=1}^{n}x_{i}^{r}e^{tx_{i}}P\left ( X=x_{i} \right )\\
\\
\Rightarrow & \frac{\mathrm{d}^{r} M_{x}\left ( t \right )}{\mathrm{d} t^{r}}\bigg{|}_{t=0} =\sum_{i=1}^{n}x_{i}^{r}P\left ( X=x_{i} \right )=\mu_{r}^{'}
\end{aligned} 	
\end{equation}$$

(2) 連續型:

$$M_{x}\left ( t \right )=\int_{-\infty }^{\infty }e^{tx}f\left (x\right )dx$$

r 階原點動差計算:

$$\begin{equation} 
\begin{aligned}
&\frac{\mathrm{d}^{r} M_{x}\left ( t \right )}{\mathrm{d} t^{r}}=\int_{-\infty }^{\infty }x^{r}e^{tx}f\left (x\right )dx\\
\\
\Rightarrow & \frac{\mathrm{d}^{r} M_{x}\left ( t \right )}{\mathrm{d} t^{r}}\bigg{|}_{t=0} =\int_{-\infty }^{\infty }x^{r}f\left (x\right )dx=\mu_{r}^{'}
\end{aligned} 	
\end{equation}$$

#### 3. 動差生成函數(moment-generating function)計算期望值與變異數

例子:

(1) 二項分配

$$M_{x}\left ( t \right )=\sum_{x=0}^{n}e^{tx}C_{x}^{n}p^{x}q^{n-x}=\sum_{x=0}^{n}C_{x}^{n}\left ( e^{t}p \right )^{x}q^{n-x}=\left(e^{t}p +q\right)^{n}$$

期望值:

$$\begin{equation} 
\begin{aligned}
&\frac{\mathrm{d} M_{x}\left ( t \right )}{\mathrm{d} t}=n\left ( e^{t}p+q \right )^{n-1}pe^{t}\\
\\
\Rightarrow & \frac{\mathrm{d} M_{x}\left ( t \right )}{\mathrm{d} t}\bigg{|}_{t=0}=n\left ( p+q \right )^{n-1}p=np
\end{aligned} 	
\end{equation}$$

變異數:

$$\begin{equation} 
\begin{aligned}
&\frac{\mathrm{d}^{2} M_{x}\left ( t \right )}{\mathrm{d} t^{2}}=n\left ( n-1 \right )\left ( e^{t}p+q \right )^{n-2}p^{2}e^{2t}+n\left ( e^{t}p+q \right )^{n-1}pe^{t}\\
\\
\Rightarrow & \frac{\mathrm{d}^{2} M_{x}\left ( t \right )}{\mathrm{d} t^{2}}\bigg{|}_{t=0}=n\left ( n-1 \right )\left ( p+q \right )^{n-2}p^{2}+n\left ( p+q \right )^{n-1}p=n\left ( n-1 \right )p^{2}+np\\
\\
\therefore & Var\left ( x \right )=n\left ( n-1 \right )p^{2}+np-\left ( np \right )^{2}=npq
\end{aligned} 	
\end{equation}$$

(2) Poisson 分配

$$M_{x}\left ( t \right )=\sum_{x=0}^{n}e^{tx}\frac{e^{-\lambda}\lambda^{x}}{x!}=e^{-\lambda}\sum_{x=0}^{n}\frac{e^{tx}\lambda^{x}}{x!}=e^{-\lambda}\sum_{x=0}^{n}\frac{\left(e^{t}\lambda\right)^{x}}{x!}=e^{-\lambda}e^{e^{t}\lambda}=e^{\lambda\left(e^{t}-1\right)}$$

期望值:

$$\begin{equation} 
\begin{aligned}
&\frac{\mathrm{d} M_{x}\left ( t \right )}{\mathrm{d} t}=\lambda e^{t}\cdot e^{\lambda\left(e^{t}-1\right)}\\
\\
\Rightarrow & \frac{\mathrm{d} M_{x}\left ( t \right )}{\mathrm{d} t}\bigg{|}_{t=0}=\lambda
\end{aligned} 	
\end{equation}$$

變異數:

$$\begin{equation} 
\begin{aligned}
&\frac{\mathrm{d}^{2} M_{x}\left ( t \right )}{\mathrm{d} t^{2}}=\lambda e^{t}\cdot e^{\lambda\left(e^{t}-1\right)}+\left ( \lambda e^{t} \right )^{2}\cdot e^{\lambda\left(e^{t}-1\right)}\\
\\
\Rightarrow & \frac{\mathrm{d}^{2} M_{x}\left ( t \right )}{\mathrm{d} t^{2}}\bigg{|}_{t=0}=\lambda+\lambda^{2}\\
\\
\therefore & Var\left ( x \right )=\lambda+\lambda^{2}-\lambda^{2}=\lambda
\end{aligned} 	
\end{equation}$$

#### 4. 動差生成函數(moment-generating function)的性質

(1) 若兩個隨機變數 $X$、$Y$ 的動差生成函數相同 $E(e^{tX})=E(e^{tY})$，則這兩個隨機變數必然相同

(2) 多個隨機變數之和

假設 $X$ 與 $Y$ 是兩個獨立的隨機變數且 $Z=X+Y$，則 $Z$ 之動差生成函數為 $M_{Z}\left ( t \right )=M_{X}\left ( t \right )\cdot M_{Y}\left ( t \right )$

$$M_{Z}\left(t\right)=E\left(e^{t\left(X+Y\right)}\right)=E\left( e^{tX}\right)\cdot E\left(e^{tY}\right)=M_{X}\left ( t \right )\cdot M_{Y}\left ( t \right )$$

(3) 隨機變數的線性變換

設 $X$、$Y$ 均為隨機變數，若 $Y=a\cdot X + b$，則 $Y$ 之動差生成函數為 $M_{Y}\left ( t \right )=e^{bt}\cdot M_{X}\left ( at \right )$

$$M_{Y}\left ( t \right )=E\left ( e^{t\left ( aX+b \right )} \right )=E\left ( e^{atX}\cdot e^{tb} \right )=E\left ( e^{atX} \right )\cdot E\left ( e^{tb} \right )=M_{X}\left ( at \right )\cdot e^{bt}$$

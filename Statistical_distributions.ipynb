## Statistical distributions

### 離散型機率分配

期望值:

$$E\left ( X \right )=\sum_{i=1}^{n}x_{i}\cdot P\left ( X=x_{i} \right )=\mu$$

變異數:

$$\begin{equation} 
\begin{aligned}
Var\left ( X \right )&=\sum_{i=1}^{n}\left ( x_{i}-\mu  \right )^{2}\cdot P\left ( X=x_{i} \right )\\
&=\sum_{i=1}^{n}\left [ x_{i}^{2}P\left ( X=x_{i} \right )-2x_{i}\mu P\left ( X=x_{i} \right )+\mu^{2}P\left ( X=x_{i} \right ) \right ]\\
&=\sum_{i=1}^{n}x_{i}^{2}P\left ( X=x_{i} \right )-2\mu \sum_{i=1}^{n}x_{i} P\left ( X=x_{i} \right )+\mu^{2}\\
&=\sum_{i=1}^{n}x^{2}P\left ( X=x_{i} \right )-\mu^{2}=E\left ( x^{2} \right )-\left [ E\left ( x \right ) \right ]^{2}
\end{aligned} 	
\end{equation}$$

#### 1. 幾何分配

(1) 定義:

已知一事件在 Bernoulli 試驗中每次成功的機率為 $p$，直到得到一次成功停止，所需試驗次數 $X$ 的機率分佈為**幾何分配**

(2) 幾何分配的機率分配函數:

$$P\left ( X=x \right )=p\cdot q^{x-1}$$

(3) 幾何分配的期望值:

$$E\left ( x \right )=\frac{1}{p}$$

(4) 幾何分配的變異數:

$$Var\left ( x \right )=\frac{q}{p^{2}}$$

(5) 幾何分配的**無記憶性質(memoryless property)**

若一隨機變數的條件機率遵循下式:

$$P\left ( T>a+b\ |\ T>a \right )=P\left ( T>b \right );\ a>0,\ b>0$$

則稱此隨機變數的機率分配具有**無記憶性質(memoryless property)**。

$$P\left ( X>x \right )=\sum_{t=x+1}^{\infty }\left ( 1-p \right )^{t-1}p=\left ( 1-p \right )^{x}$$

$$\begin{equation} 
\begin{aligned}
\Rightarrow P\left ( T>a+b|T>a \right )=\frac{P\left ( T>a+b \right )}{P\left ( T>a \right )}=\frac{\left ( 1-p \right )^{a+b}}{\left ( 1-p \right )^{a}}=\left ( 1-p \right )^{b}=P\left ( T>b \right )
\end{aligned} 	
\end{equation}$$

#### 2. 負二項分配

(1) 定義:

給定一正整數 k，持續進行 Bernoulli 試驗，直到得到第 k 次成功為止，若總共試驗的次數為 $X$，則 $X$ 的機率分佈為**負二項分配**

(2) 負二項分配的機率分配函數:

$$P\left ( X=x \right )=C_{k-1}^{x-1}p^{k-1}q^{x-k}\cdot p=C_{k-1}^{x-1}p^{k}q^{x-k}$$

(3) 負二項分配的期望值:

$$E\left ( x \right )=\frac{k}{p}$$

(4) 負二項分配的變異數:

$$Var\left ( x \right )=\frac{kq}{p^{2}}$$

#### 3. 超幾何分配

(1) 定義:

設隨機變數 $X$ 為二項試驗，但每次自母體(總個數 N，成功的母體數目: k 個)抽出的樣本(總個數 n)**不放回**，則 $X$ 的機率分配為**超幾何分配**

(2) 超幾何分配的機率分配函數:

$$P\left ( X=x \right )=h\left ( x; N, n, k \right )=\frac{C_{x}^{k}\cdot C_{n-x}^{N-k}}{C_{n}^{N}}$$

(3) 超幾何分配的期望值:

$$\begin{equation} 
\begin{aligned}
E\left ( x \right )&=\sum_{x=0}^{n}x\cdot \frac{C_{x}^{k}\cdot C_{n-x}^{N-k}}{C_{n}^{N}}=\sum_{x=0}^{n}x\cdot \frac{k!\cdot  C_{n-x}^{N-k}}{x!\left ( k-x \right )!\cdot C_{n}^{N}}\\
\\
&=\sum_{x=1}^{n}\frac{k!\cdot  C_{n-x}^{N-k}}{\left ( x-1 \right )!\left ( k-x \right )!\cdot C_{n}^{N}}=k\sum_{x=1}^{n}\frac{\left ( k-1 \right )!\cdot  C_{n-x}^{N-k}}{\left ( x-1 \right )!\left ( k-x \right )!\cdot C_{n}^{N}}\\
\\
&=k\sum_{x=1}^{n} \frac{C_{x-1}^{k-1}\cdot C_{n-x}^{N-k}}{C_{n}^{N}},\ (令\ y=x-1)\\
\\
&=k\sum_{y=0}^{n-1} \frac{C_{y}^{k-1}\cdot C_{n-1-y}^{N-k}}{C_{n}^{N}}=\frac{nk}{N}\sum_{y=0}^{n-1} \frac{C_{y}^{k-1}\cdot C_{n-1-y}^{\left ( N-1 \right )-\left ( k-1 \right )}}{C_{n-1}^{N-1}}=\frac{nk}{N}=np
\end{aligned} 	
\end{equation}$$

其中

$$C_{n}^{N}=\frac{N!}{n!\left ( N-n \right )!}=\frac{N}{n}\frac{\left ( N-1 \right )!}{\left ( n-1 \right )!\left ( N-n \right )!}=\frac{N}{n}C_{n-1}^{N-1}$$

(4) 超幾何分配的變異數:

$$\begin{equation} 
\begin{aligned}
E\left [ x\left ( x-1 \right ) \right ]&=\sum_{x=0}^{n}x\left ( x-1 \right )\cdot \frac{C_{x}^{k}\cdot C_{n-x}^{N-k}}{C_{n}^{N}}=\sum_{x=0}^{n}x\left ( x-1 \right )\cdot \frac{k!\cdot  C_{n-x}^{N-k}}{x!\left ( k-x \right )!\cdot C_{n}^{N}}\\
\\
&=\sum_{x=2}^{n}\frac{k!\cdot  C_{n-x}^{N-k}}{\left ( x-2 \right )!\left ( k-x \right )!\cdot C_{n}^{N}}=k\left ( k-1 \right )\sum_{x=2}^{n}\frac{\left ( k-2 \right )!\cdot  C_{n-x}^{N-k}}{\left ( x-2 \right )!\left ( k-x \right )!\cdot C_{n}^{N}}\\
\\
&=k\left ( k-1 \right )\sum_{x=2}^{n} \frac{C_{x-2}^{k-2}\cdot C_{n-x}^{N-k}}{C_{n}^{N}},\ (令\ y=x-2)\\
\\
&=k\left ( k-1 \right )\sum_{y=0}^{n-2} \frac{C_{y}^{k-2}\cdot C_{n-2-y}^{N-k}}{C_{n}^{N}}=\frac{n\left ( n-1 \right )k\left ( k-1 \right )}{N\left ( N-1 \right )}\sum_{y=0}^{n-2} \frac{C_{y}^{k-2}\cdot C_{n-2-y}^{\left ( N-2 \right )-\left ( k-2 \right )}}{C_{n-2}^{N-2}}\\
\\
&=\frac{n\left ( n-1 \right )k\left ( k-1 \right )}{N\left ( N-1 \right )}\\
\\
\therefore  Var\left ( x \right )&=E\left ( x^{2} \right )-\mu^{2}=E\left [ x\left ( x-1 \right )+x \right ]-\mu^{2}=E\left [ x\left ( x-1 \right ) \right ]+\mu-\mu^{2}\\
\\
&=\frac{n\left ( n-1 \right )k\left ( k-1 \right )}{N\left ( N-1 \right )}+\frac{nk}{N}-\frac{n^{2}k^{2}}{N^{2}}=\frac{nk\left ( N-n \right )\left ( N-k \right )}{N^{2}\left ( N-1 \right )}\\
\\
&=\frac{\left ( N-n \right )}{\left ( N-1 \right )}\cdot n\cdot \frac{k}{N}\cdot \left ( 1-\frac{k}{N} \right )=\frac{\left ( N-n \right )}{\left ( N-1 \right )}\cdot npq
\end{aligned} 	
\end{equation}$$

其中 $\left ( N-n \right )/\left ( N-1 \right )$ 稱為**有限母體校正因子**，當 $n<<N$ 時，超幾何分配會接近**二項分配**。 $(n/N < 0.05)$

#### 4. 二項分配

(1) 二項試驗(Bernoulli 試驗):

在進行 n 次的的試驗中，每次試驗只有兩種可能(成功或失敗)，每次成功 $p$ (失敗 $q$) 的機率都相同(母體總數 N，每次抽樣後**均放回**)，且每次試驗皆互相獨立。

(2) 二項分配的機率分配函數:

假設實驗 n 次，每次成功機率為 $p$，失敗機率為 $q\ (=1-p)$，$X$ 表示成功的次數，則機率分配函數為

$$P\left ( X=x \right )=b\left ( x; n,p \right )=C_{x}^{n}p^{x}q^{n-x}$$

(3) 二項分配的期望值:

$$\begin{equation} 
\begin{aligned}
E\left ( x \right )&=\sum_{x=0}^{n}x\cdot C_{x}^{n}p^{x}q^{n-x}=\sum_{x=1}^{n}x\cdot \frac{n!}{x!\left ( n-x \right )!}p^{x}q^{n-x}\\
&=np\cdot \sum_{x=1}^{n}\frac{\left ( n-1 \right )!}{\left ( n-1 \right )!\left ( n-x \right )!}p^{x-1}q^{n-x}=np\cdot\sum_{x=0}^{n} C_{x-1}^{n-1}p^{x-1}q^{n-x}\\
&=np\cdot \left ( p+q \right )^{n}=np
\end{aligned} 	
\end{equation}$$

(4) 二項分配的變異數:

$$\begin{equation} 
\begin{aligned}
Var\left ( x \right )&=\sum_{x=0}^{n}x^{2}\cdot C_{x}^{n}p^{x}q^{n-x}-\left ( np \right )^{2}=\sum_{x=0}^{n}\left [ x\left ( x-1 \right )+x \right ]\cdot C_{x}^{n}p^{x}q^{n-x}-n^{2}p^{2}\\
&=\sum_{x=0}^{n}x\left ( x-1 \right )\cdot C_{x}^{n}p^{x}q^{n-x}+np-n^{2}p^{2}\\
&=\sum_{x=0}^{n}x\left ( x-1 \right )\cdot \frac{n!}{x!\left ( n-x \right )!}p^{x}q^{n-x}+np-n^{2}p^{2}\\
&=\sum_{x=0}^{n}\frac{n!}{\left ( x-2 \right )!\left ( n-x \right )!}p^{x}q^{n-x}+np-n^{2}p^{2}\\
&=n\left ( n-1 \right )p^{2}\sum_{x=0}^{n}\frac{\left ( n-2 \right )!}{\left ( x-2 \right )!\left ( n-x \right )!}p^{x-2}q^{n-x}+np-n^{2}p^{2}\\
&=n\left ( n-1 \right )p^{2}+np-n^{2}p^{2}=-np^{2}+np=np\left ( 1-p \right )=npq
\end{aligned} 	
\end{equation}$$

#### 5. Poisson 分配 $\Rightarrow$ 探討一段時間內發生事件的次數 $\lambda$

(1) Poisson 分配的機率分配函數:

$$P\left ( X=x \right )=\frac{e^{-\lambda }\lambda^{x}}{x!}$$

(2) Poisson 分配的性質:

a. 一段時間內發生的次數與另一段時間內發生的次數獨立

b. 一段時間內發生的平均次數與時間的長短成比例

c. 在極短的時間內發生的機率趨近零

(3) 由**二項分配**推導 Poisson 分配的機率分配函數

$n\rightarrow \infty ,\ p\rightarrow 0$ 的二項分配: ($np=\lambda$)

$$\begin{equation} 
\begin{aligned}
&P\left ( X=x \right )=C_{x}^{n}p^{x}q^{n-x}=\frac{n!}{x!\left ( n-x \right )!}p^{x}q^{n-x}=\frac{n\left ( n-1 \right )\cdot \cdot \cdot \left ( n-x+1 \right )}{x!}\cdot \left ( \frac{\lambda}{n} \right )^{x}\cdot \left ( 1-\frac{\lambda}{n} \right )^{n-x}\\
\\
\Rightarrow &\lim_{n\rightarrow \infty }P\left ( X=x \right )=\lim_{n\rightarrow \infty }\frac{n\left ( n-1 \right )\cdot \cdot \cdot \left ( n-x+1 \right )}{n^{x}}\cdot \frac{\lambda^{x}}{x!}\cdot \left ( 1-\frac{\lambda}{n} \right )^{n}\cdot \left ( 1-\frac{\lambda}{n} \right )^{-x}=\frac{e^{-\lambda }\lambda^{x}}{x!}
\end{aligned} 	
\end{equation}$$

Note:

$$\lim_{n\rightarrow \infty }\frac{n\left ( n-1 \right )\cdot \cdot \cdot \left ( n-x+1 \right )}{n^{x}}=1$$

$$\lim_{n\rightarrow \infty }\left ( 1-\frac{\lambda}{n} \right )^{n}=e^{-\lambda }$$

$$\lim_{n\rightarrow \infty }\left ( 1-\frac{\lambda}{n} \right )^{-x}=1$$

(4) 由 Poisson 分配近似**二項分配**:

當 $n$ 很大、 $p$ 很小且 $\lambda=np<5$ 時，**二項分配**可用 Poisson 分配計算

### 連續型機率分配

期望值:

$$E\left ( x \right )=\int_{-\infty }^{\infty }x\cdot f\left ( x \right )dx=\mu$$

變異數:

$$\begin{equation} 
\begin{aligned}
Var\left ( x \right )&=\int_{-\infty }^{\infty }\left ( x-\mu  \right )^{2}\cdot f\left ( x \right )dx\\
\\
&=\int_{-\infty }^{\infty }\left [ x^{2}f\left ( x \right )-2x\mu f\left ( x \right )+\mu^{2}f\left ( x \right ) \right ]dx\\
\\
&=\int_{-\infty }^{\infty }x^{2}f\left ( x \right )dx-2\mu \int_{-\infty }^{\infty }x\cdot f\left ( x \right )dx+\mu^{2}\\
\\
&=\int_{-\infty }^{\infty }x^{2}f\left ( x \right )dx-\mu^{2}=E\left ( x^{2} \right )-\left [ E\left ( x \right ) \right ]^{2}
\end{aligned} 	
\end{equation}$$

#### 1. Gamma 分配

(1) Gamma 函數的定義:

$$\Gamma \left ( \alpha  \right )=\int_{0}^{\infty }x^{\alpha-1}e^{-x}dx;\ \alpha>0$$

(2) Gamma 函數的性質:

(i) $\Gamma \left ( \alpha  \right )=\left ( \alpha-1  \right )\Gamma \left ( \alpha-1  \right )=\left ( \alpha-1  \right )!$

(ii) $\Gamma \left ( \frac{1}{2}  \right )=\sqrt{\pi }$

(3) Gamma 分配的機率密度函數:

$$f\left ( x \right )=\frac{1}{\beta^{\alpha }\Gamma \left(\alpha\right )}x^{\alpha-1}e^{-\frac{x}{\beta }};\ \alpha>0, \beta>0$$

(4) Gamma 分配的參數說明:

- $\alpha$: 形狀參數，影響機率密度函數的**陡峭程度**

- $\beta$: 尺度參數，影響機率密度函數的**散佈程度**

(5) Gamma 分配的期望值:

$$\begin{equation} 
\begin{aligned}
\mu_{r}^{'}&=E\left ( x^{r} \right )=\frac{1}{\beta^{\alpha }\Gamma \left(\alpha\right )}\int_{-\infty }^{\infty }x^{r+\alpha-1}e^{-\frac{x}{\beta }}dx\ \ (令\ y=\frac{x}{\beta })\\
\\
&=\frac{\beta^{r+\alpha-1}}{\beta^{\alpha }\Gamma \left(\alpha\right )}\int_{-\infty }^{\infty }y^{r+\alpha-1}e^{-y}\beta dy=\frac{\beta^{r}}{\Gamma \left(\alpha\right )}\int_{-\infty }^{\infty }y^{r+\alpha-1}e^{-y} dy\\
\\
&=\frac{\beta^{r}}{\Gamma \left(\alpha\right )}\Gamma \left(r+\alpha\right )\\
\\
\Rightarrow \mu&=\mu_{1}^{'}=\frac{\beta}{\Gamma \left(\alpha\right )}\Gamma \left(1+\alpha\right )=\alpha \beta
\end{aligned} 	
\end{equation}$$

(6) Gamma 分配的變異數:

$$\begin{equation} 
\begin{aligned}
Var\left ( x \right )&=E\left ( x^{2} \right )-\mu^{2}=\mu_{2}^{'}-\mu^{2}\\
\\
&=\frac{\beta^{2}}{\Gamma \left(\alpha\right )}\Gamma \left(2+\alpha\right )-\left(\alpha \beta \right )^{2}\\
\\
&=\beta^{2}\alpha \left(\alpha+1\right )-\alpha^{2}\beta^{2}=\alpha\beta^{2}
\end{aligned} 	
\end{equation}$$

#### 2. 指數分配 ($\alpha=1$ 的 Gamma 分配)

(1) 指數分配的機率密度函數:

$$\begin{equation} 
\begin{aligned}
f\left ( x \right )&=\frac{1}{\beta}e^{-\frac{x}{\beta }};\ \beta>0\\
\\
&=\lambda e^{-\lambda x};\ \lambda>0
\end{aligned} 	
\end{equation}$$

(2) 指數分配的期望值:

$$\begin{equation} 
\begin{aligned}
\mu_{r}^{'}&=E\left ( x^{r} \right )=\frac{1}{\beta}\int_{-\infty }^{\infty }x^{r}e^{-\frac{x}{\beta }}dx\ \ (令\ y=\frac{x}{\beta })\\
\\
&=\beta^{r}\int_{-\infty }^{\infty }y^{r}e^{-y} dy=\beta^{r}\Gamma \left(r+1\right )\\
\\
\Rightarrow \mu&=\mu_{1}^{'}=\beta\ \Gamma \left(2\right )=\beta
\end{aligned} 	
\end{equation}$$

(3) 指數分配的變異數:

$$\begin{equation} 
\begin{aligned}
Var\left ( x \right )&=E\left ( x^{2} \right )-\mu^{2}=\mu_{2}^{'}-\mu^{2}\\
\\
&=\beta^{2}\Gamma \left(3\right )-\beta^{2}=2\beta^{2}-\beta^{2}=\beta^{2}
\end{aligned} 	
\end{equation}$$

(4) 指數分配的用途: 被使用在 "等待時間" 或 "壽命" 的模式

例子:

- 前後兩顧客到服務台之中間間隔的時間

- 電子零組件的壽命問題

(5) 指數分配的機率計算:

$$P\left ( X>a \right )=\int_{a}^{\infty}\frac{1}{\beta}e^{-\frac{x}{\beta}}dx=-e^{-\frac{x}{\beta }}\bigg{|}_{a}^{\infty }=e^{-\frac{a}{\beta }}$$

(6) 指數分配的應用: 排隊理論

(7) 指數分配與** Poisson 分配**的關係:

在 Poisson 實驗下，若每單位時間發生某事件的平均次數為 $\lambda$ 次，則兩相臨事件間隔的時間服從**指數分配**，其平均值為 $\beta$，其中 $\beta=1/\lambda$

(8) 指數分配的**無記憶性質(memoryless property)**:

$$\begin{equation} 
\begin{aligned}
P\left ( T>a+b|T>a \right )&=\frac{P\left ( T>a+b \right )}{P\left ( T>a \right )}=\frac{\int_{a+b}^{\infty}\lambda e^{-\lambda x}dx}{\int_{a}^{\infty}\lambda e^{-\lambda x}dx}=\frac{-e^{-\lambda x}\bigg{|}_{a+b}^{\infty }}{-e^{-\lambda x}\bigg{|}_{a}^{\infty }}=\frac{e^{-\lambda \left ( a+b \right )}}{e^{-\lambda a}}=e^{-\lambda b}=P\left ( T>b \right )
\end{aligned} 	
\end{equation}$$

#### 3. 卡方分配 ($\alpha=\nu/2,\ \beta=2$ 的 Gamma 分配)

(1) 卡方分配(chi-square distribution, $\chi^{2}$)的機率密度函數:

$$f\left ( x \right )=\frac{1}{2^{\frac{\nu}{2}}\Gamma \left(\frac{\nu}{2}\right )}x^{\frac{\nu}{2}-1}e^{-\frac{x}{2}};\ x>0$$

其中自由度 $\nu$ 為正整數。 

(2) 卡方分配的動差生成函數:

$$\begin{equation} 
\begin{aligned}
M_{x}\left ( t \right )&=\int_{0}^{\infty }e^{tx}\frac{1}{2^{\frac{\nu}{2}}\Gamma \left(\frac{\nu}{2}\right )}x^{\frac{\nu}{2}-1}e^{-\frac{x}{2}}dx,\ (令\ y=x\left ( 1-2t \right )/2)\\
\\
&=\int_{0}^{\infty }\frac{1}{2^{\frac{\nu}{2}}\Gamma \left(\frac{\nu}{2}\right )}x^{\frac{\nu}{2}-1}e^{-x\left ( 1-2t \right )/2}dx\\
\\
&=\int_{0}^{\infty }\frac{1}{2^{\frac{\nu}{2}}\Gamma \left(\frac{\nu}{2}\right )}\cdot \left (\frac{2y}{1-2t} \right )^{\frac{\nu}{2}-1}\cdot e^{-y}\cdot \left (\frac{2}{1-2t} \right )dy\\
\\
&=\frac{1}{2^{\frac{\nu}{2}}\Gamma \left(\frac{\nu}{2}\right )}\cdot \left (\frac{2}{1-2t} \right )^{\frac{\nu}{2}}\cdot\int_{0}^{\infty }y^{\frac{\nu}{2}-1}e^{-y}dy\\
\\
&=\frac{1}{\left ( 1-2t \right )^{\frac{\nu}{2}}\Gamma \left(\frac{\nu}{2}\right )}\cdot \int_{0}^{\infty }y^{\frac{\nu}{2}-1}e^{-y}dy\\
\\
&=\frac{1}{\left ( 1-2t \right )^{\frac{\nu}{2}}\Gamma \left(\frac{\nu}{2}\right )}\cdot \Gamma \left ( \frac{\nu}{2} \right )=\left ( 1-2t \right )^{-\frac{\nu}{2}}
\end{aligned} 	
\end{equation}$$

(3) 卡方分配的期望值:

$$\begin{equation} 
\begin{aligned}
&\frac{\mathrm{d} M_{x}\left ( t \right )}{\mathrm{d} t}=-\frac{\nu}{2}\left ( 1-2t \right )^{-\frac{\nu}{2}-1}\cdot \left ( -2 \right )=\nu \left ( 1-2t \right )^{-\frac{\nu}{2}-1}\\
\\
\Rightarrow & \frac{\mathrm{d} M_{x}\left ( t \right )}{\mathrm{d} t}\bigg{|}_{t=0}=\nu
\end{aligned} 	
\end{equation}$$

(4) 卡方分配的變異數:

$$\begin{equation} 
\begin{aligned}
&\frac{\mathrm{d}^{2} M_{x}\left ( t \right )}{\mathrm{d} t^{2}}=\nu \left ( -\frac{\nu}{2}-1 \right )\left ( 1-2t \right )^{-\frac{\nu}{2}-2}\cdot \left ( -2 \right )=\nu \left ( \nu +2 \right )\left ( 1-2t \right )^{-\frac{\nu}{2}-2}\\
\\
\Rightarrow & \frac{\mathrm{d}^{2} M_{x}\left ( t \right )}{\mathrm{d} t^{2}}\bigg{|}_{t=0}=\nu \left ( \nu +2 \right )=\mu_{2}^{'}\\
\\
\therefore & Var\left ( x \right )=\mu_{2}^{'}-\mu^{2}=\nu \left ( \nu +2 \right )-\nu^{2}=2\nu
\end{aligned} 	
\end{equation}$$

(5) 卡方分配的性質

(i) 當自由度 $\nu$ 趨近於無限大時，卡方分配會趨近於**常態分配**

(ii) 設 $X_{1}$ 與 $X_{2}$ 為兩相互獨立的卡方分配，其自由度分別為 $\nu_{1}$ 和 $\nu_{2}$，則 $Y=X_{1}+X_{2}$ 為自由度 $\nu_{1} + \nu_{2}$ 的卡方分配

$$\begin{equation} 
\begin{aligned}
&M_{X_{1}}\left ( t \right )=\left ( 1-2t \right )^{-\frac{\nu_{1}}{2}},\ M_{X_{2}}\left ( t \right )=\left ( 1-2t \right )^{-\frac{\nu_{2}}{2}}\\
\\
\Rightarrow & M_{Y}\left ( t \right )=M_{X_{1}}\left ( t \right )\cdot M_{X_{2}}\left ( t \right )=\left ( 1-2t \right )^{-\frac{\nu_{1}}{2}}\cdot \left ( 1-2t \right )^{-\frac{\nu_{2}}{2}}=\left ( 1-2t \right )^{-\frac{1}{2}\left ( \nu_{1}+\nu_{2} \right )}
\end{aligned} 	
\end{equation}$$

#### 4. 常態分配

(1) 常態分配的機率密度函數: $N(\mu, \sigma^{2})$

$$f\left ( x \right )=\frac{1}{\sqrt{2\pi}\sigma }e^{-\frac{1}{2}\left ( \frac{x-\mu}{\sigma}\right)^{2}}$$

(2) 常態分配的動差生成函數:

$$\begin{equation} 
\begin{aligned}
M_{x}\left ( t \right )&=\int_{-\infty}^{\infty}e^{tx}\cdot \frac{1}{\sqrt{2\pi}\sigma }e^{-\frac{1}{2}\left ( \frac{x-\mu}{\sigma}\right)^{2}}dx\\
\\
&=\int_{-\infty}^{\infty}\frac{1}{\sqrt{2\pi}\sigma }e^{-\frac{\left [ x-\left ( \mu+\sigma ^{2}t \right )^{2} \right ]-2\mu \sigma^{2}t-\sigma^{4}t^{2}}{2\sigma^{2}}}dx\\
\\
&=e^{\left ( 2\mu t+\sigma^{2}t^{2} \right )/2}\cdot \int_{-\infty}^{\infty}\frac{1}{\sqrt{2\pi}\sigma }e^{-\frac{x-\left ( \mu+\sigma ^{2}t \right )^{2}}{2\sigma^{2}}}dx,\ \ (令\ y=\frac{x-\left ( \mu+\sigma ^{2}t \right )}{\sigma})\\
\\
&=e^{\left ( 2\mu t+\sigma^{2}t^{2} \right )/2}\cdot \int_{-\infty}^{\infty}\frac{1}{\sqrt{2\pi}}e^{-\frac{y^{2}}{2}}dy=e^{\left ( 2\mu t+\sigma^{2}t^{2} \right )/2}
\end{aligned} 	
\end{equation}$$

(3) 常態分配的期望值:

$$\begin{equation} 
\begin{aligned}
&\frac{\mathrm{d} M_{x}\left ( t \right )}{\mathrm{d} t}=e^{\left ( 2\mu t+\sigma^{2}t^{2} \right )/2}\cdot \left (\mu+\sigma^{2}t\right )\\
\\
\Rightarrow & \frac{\mathrm{d} M_{x}\left ( t \right )}{\mathrm{d} t}\bigg{|}_{t=0}=\mu
\end{aligned} 	
\end{equation}$$

(4) 常態分配的變異數:

$$\begin{equation} 
\begin{aligned}
&\frac{\mathrm{d}^{2} M_{x}\left ( t \right )}{\mathrm{d} t^{2}}=e^{\left ( 2\mu t+\sigma^{2}t^{2} \right )/2}\cdot \left (\mu+\sigma^{2}t\right )^{2}+e^{\left ( 2\mu t+\sigma^{2}t^{2} \right )/2}\cdot \sigma^{2}\\
\\
\Rightarrow & \frac{\mathrm{d}^{2} M_{x}\left ( t \right )}{\mathrm{d} t^{2}}\bigg{|}_{t=0}=\mu^{2}+\sigma^{2}\\
\\
\therefore & Var\left ( x \right )=\mu_{2}^{'}-\mu^{2}=\mu^{2}+\sigma^{2}-\mu^{2}=\sigma^{2}
\end{aligned} 	
\end{equation}$$
(5) 常態分配與**二項分配**的關係:

二項分配中，當試驗次數 n 越大時，無論成功機率 P 值為何，其分配會慢慢呈現對稱的圖形。在 $np>5$ 且 $nq>5$ 時，可以用**常態分配**估計機率。

以常態分配估計二項分配機率的連續校正因子(correction for continuity):

- $P\left ( a\leq x\leq b \right )\overset{\cdot}{=}P\left ( a-0.5<x<b+0.5 \right )$

- $P\left ( x \leq a \right )\overset{\cdot}{=}P\left ( x<a+0.5 \right )$

- $P\left ( x \geq a \right )\overset{\cdot}{=}P\left ( x>a-0.5 \right )$

(6) 標準常態分配: $Z$ ~ $N(0, 1)$

機率密度函數:

$$f\left ( z \right )=\frac{1}{\sqrt{2\pi }}e^{-\frac{z^{2}}{2}}$$

標準常態分配的動差生成函數:

$$M_{z}\left ( t \right )=e^{t^{2}/2}$$

常態分配的標準化:

若隨機變數 $X$ 服從常態分配，其平均數為 $\mu$，標準差為 $\sigma$，機率 $P(a<X<b)$ 可用隨機變數 $Z=(X-\mu)/\sigma$ 進行標準化

$$\begin{equation}
\begin{aligned}
P\left ( a<x<b \right )&=\int_{a}^{b}\frac{1}{\sqrt{2\pi}\sigma }e^{-\frac{1}{2}\left ( \frac{x-\mu}{\sigma}\right)^{2}}dx=\int_{\frac{a-\mu}{\sigma}}^{\frac{b-\mu}{\sigma}}\frac{1}{\sqrt{2\pi}\sigma }e^{-\frac{z^{2}}{2}}\sigma dz=\int_{\frac{a-\mu}{\sigma}}^{\frac{b-\mu}{\sigma}}\frac{1}{\sqrt{2\pi}}e^{-\frac{z^{2}}{2}}dz\\
\\
&=P\left ( \frac{a-\mu}{\sigma}<z<\frac{b-\mu}{\sigma}\right )
\end{aligned}
\end{equation}$$

(7) 標準常態分配之隨機變數的平方 $Z^{2}$ 為 自由度=1 的卡方分配: $Z^{2}$ ~ $\chi_{1}^{2}$

若隨機變數 $X$ 服從常態分配，其平均數為 $\mu$，標準差為 $\sigma$，則隨機變數 $Y=Z^{2}=(X-\mu)^{2}/\sigma^{2}$ 服從**自由度為 1 的卡方分配**

$$Z=\frac{X-\mu}{\sigma};\ \ f\left ( z \right )=\frac{1}{\sqrt{2\pi }}e^{-\frac{z^{2}}{2}},\ \ -\infty <z<\infty$$

$$y=z^{2}\Rightarrow z_{1}=-\sqrt{y},\ z_{2}=\sqrt{y}$$

$$\Rightarrow \left\{\begin{matrix}
\begin{aligned}
J_{1}&=\frac{\mathrm{d} z_{1}}{\mathrm{d} y}=-\frac{1}{2\sqrt{y}}\\ 
J_{2}&=\frac{\mathrm{d} z_{2}}{\mathrm{d} y}=\frac{1}{2\sqrt{y}}
\end{aligned}
\end{matrix}\right.$$

$$\begin{equation}
\begin{aligned}
g\left ( y \right )&=\frac{1}{\sqrt{2\pi }}e^{-\frac{y}{2}}\big{|}J_{1}\big{|}+\frac{1}{\sqrt{2\pi }}e^{-\frac{y}{2}}\big{|}J_{2}\big{|}\\
&=\frac{1}{\sqrt{2\pi }}e^{-\frac{y}{2}}\bigg{|}-\frac{1}{2\sqrt{y}}\bigg{|}+\frac{1}{\sqrt{2\pi }}e^{-\frac{y}{2}}\bigg{|}\frac{1}{2\sqrt{y}}\bigg{|}\\
&=\frac{1}{\sqrt{2\pi }}y^{-\frac{1}{2}}e^{-\frac{y}{2}},\ y>0
\end{aligned}
\end{equation}$$

$\nu=1$ 時，卡方分配(chi-square distribution, $\chi^{2}$)的機率密度函數:

$$f\left ( y \right )=\frac{1}{2^{\frac{1}{2}}\Gamma \left(\frac{1}{2}\right )}y^{-\frac{1}{2}}e^{-\frac{y}{2}}=\frac{1}{\sqrt{2\pi}}y^{-\frac{1}{2}}e^{-\frac{y}{2}} ;\ y>0$$

(8) 若隨機變數 $X_{1}, X_{2},...,X_{n}$ 服從常態分配，其平均數為 $\mu$，標準差為 $\sigma$，則 $\sum_{i=1}^{n}\left ( \frac{X_{i}-\mu}{\sigma} \right )^{2}$ 為 **自由度 $=n$ 的卡方分配**、$\left ( n-1 \right )s^{2}/\sigma^{2}$ 為 **自由度 $=n-1$ 的卡方分配**

$$\begin{equation}
\begin{aligned}
\sum_{i=1}^{n}\left ( \frac{X_{i}-\mu}{\sigma} \right )^{2}&=\sum_{i=1}^{n}\left ( \frac{X_{i}-\bar{X}+\bar{X}-\mu}{\sigma} \right )^{2}\\
\\
&=\sum_{i=1}^{n}\left ( \frac{X_{i}-\bar{X}}{\sigma} \right )^{2}+\sum_{i=1}^{n}\left ( \frac{\bar{X}-\mu}{\sigma} \right )^{2}+2\sum_{i=1}^{n}\left ( \frac{X_{i}-\bar{X}}{\sigma}\cdot \frac{\bar{X}-\mu}{\sigma} \right )\\
\\
&=\frac{\sum_{i=1}^{n}\left ( X_{i}-\bar{X} \right )^{2}}{\sigma^{2}}+n\left ( \frac{\bar{X}-\mu}{\sigma} \right )^{2}=\frac{\sum_{i=1}^{n}\left ( X_{i}-\bar{X} \right )^{2}}{\sigma^{2}}+\left ( \frac{\bar{X}-\mu}{\sigma/\sqrt{n}} \right )^{2}\\
\\
&=\frac{\left ( n-1 \right )s^{2}}{\sigma^{2}}+\left ( \frac{\bar{X}-\mu}{\sigma/\sqrt{n}} \right )^{2}\sim \chi_{n-1}^{2}+\chi_{1}^{2}
\end{aligned}
\end{equation}$$

Note:

$$s^{2}=\frac{\sum_{i=1}^{n}\left ( X_{i}-\bar{X} \right )^{2}}{n-1}
\Rightarrow \frac{\left ( n-1 \right )s^{2}}{\sigma^{2}}=\frac{\sum_{i=1}^{n}\left ( X_{i}-\bar{X} \right )^{2}}{\sigma^{2}}$$

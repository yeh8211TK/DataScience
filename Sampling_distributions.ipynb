### 抽樣分配 (sampling distribution)

#### 一. 統計量: 樣本所顯示的特徵數據 (參數: 母體的特徵數據)

1.抽樣分配: 統計量(隨機變數)的機率分配

- $\bar{X}$ 的機率分配 $\rightarrow$ 樣本平均數的抽樣分配


- $\hat{p}$ 的機率分配 $\rightarrow$ 樣本比例的抽樣分配


- $s^2$ 的機率分配 $\rightarrow$ 樣本變異數的抽樣分配


2.標準誤(standard error of the mean): 抽樣分配的標準差。當使用樣本(sample)資料對母體(population)平均值做估計時，此估計結果的誤差程度

#### 二. 樣本平均數 ($\bar{X}$) 的抽樣分配

1.由一組大小為 N 的母體抽出一組樣本 $X_{1}、X_{2}、...、X_{n}$，樣本平均數是 

$$\bar{X} = \frac{X_{1}+X_{2}+...+X_{n}}{n}$$

則 

(1) $E(\bar{X})=\mu_{\bar{X}}=\mu$   

(2) $Var(\bar{X})=\sigma_{\bar{X}}^{2}=\sigma^{2}/n$

2.分類:

(1) 常態母體

a. 大樣本: $\bar{X} \sim N\left ( \mu , \frac{\sigma }{\sqrt{n}} \right ) \sim N\left ( \mu , \frac{s}{\sqrt{n}} \right )$

[註] 有限母體、大樣本且 $\sigma$ 未知: $\bar{X} \sim N\left ( \mu , \frac{s}{\sqrt{n}}\cdot \sqrt{\frac{N-n}{n-1}} \right )$

b. 小樣本 

(i) $\sigma$ 已知: $\bar{X} \sim N\left ( \mu , \frac{\sigma }{\sqrt{n}} \right )$

(ii) $\sigma$ 未知: $\bar{X} \sim t\left ( \mu , \frac{s}{\sqrt{n}} \right )$

(2) 非常態母體

a. 大樣本 (C.L.T): $\bar{X} \sim N\left ( \mu , \frac{\sigma }{\sqrt{n}} \right )$

b. 小樣本: 視母體狀況 (Non-parametric method)

#### 三. 樣本比例 ($\hat{p}$) 的抽樣分配

1.由一組大小為 N 的母體抽出一組樣本 $X_{1}、X_{2}、...、X_{n}$，樣本比例是 $\hat{p}=x/n$，則

(1) $E(\hat{p})=E(\frac{x}{n})=\frac{1}{n}E(x)=\frac{1}{n}np=p$

(2) $Var(\hat{p})=Var(\frac{x}{n})=\frac{1}{n^{2}}Var(x)=\frac{1}{n^{2}}npq=pq/n$

2.分類:

(1) 大樣本 ($np>5$ & $nq>5$): $\hat{p} \sim N\left ( p, \sqrt{\frac{pq}{n}} \right )$

(2) 小樣本

a. 無限母體(取後放回): $\hat{p} \sim binomial\left ( p, \sqrt{\frac{pq}{n}} \right )$

b. 有限母體(取後不放回): $\hat{p} \sim hypergeometry\left ( p, \sqrt{\frac{pq}{n}\cdot \frac{N-n}{N-1}} \right )$

#### 四. 樣本變異數 ($s^{2}$) 的抽樣分配

$$s^{2}=\frac{\sum_{i=1}^{n}\left ( X_{i}-\bar{X} \right )^{2}}{n-1}
\Rightarrow \frac{\left ( n-1 \right )s^{2}}{\sigma^{2}}=\frac{\sum_{i=1}^{n}\left ( X_{i}-\bar{X} \right )^{2}}{\sigma^{2}}\sim \chi_{n-1}^{2} (自由度為 n-1 的卡方分配)$$

#### 五. 中央極限定理(C.L.T)

假設 $X_{1}、X_{2}、...、X_{n}$ 是獨立且同分布的隨機變數，期望值為 $\mu$、變異數為 $\sigma^{2}$，當樣本數增大(n>30)時，其樣本平均數的分佈趨近於常態分佈

$$\lim_{n\rightarrow \infty }\frac{X_{1}+X_{2}+...+X_{n}}{n}\sim N\left ( \mu ,\frac{\sigma }{\sqrt{n}} \right )$$

證明:

$$S_{n}=X_{1}+X_{2}+...+X_{n}$$

$$\Rightarrow \bar{X}=\frac{X_{1}+X_{2}+...+X_{n}}{n}=\frac{S_{n}}{n}$$

$$M_{S_{n}}\left ( t \right )=\left [ M_{X}\left ( t \right ) \right ]^{n}$$

$$\Rightarrow M_{\bar{X}}\left ( t \right )=M_{S_{n}}\left ( \frac{t}{n} \right )=\left [ M_{X}\left ( \frac{t}{n} \right ) \right ]^{n}$$

[Note 1: X 與 Y 是兩個獨立的隨機變數且 $Z=X+Y$，則 $M_{Z}\left ( t \right )=M_{X}\left ( t \right )\cdot M_{Y}\left ( t \right )$]

[Note 2: X、Y 均為隨機變數且 $Y=a\cdot X + b$，則 $M_{Y}\left ( t \right )=e^{bt}\cdot M_{X}\left ( at \right )$]

$$\begin{equation} 
\begin{aligned}
Y&=\frac{\bar{X}-\mu}{\sigma/\sqrt{n}}=\frac{\sqrt{n}}{\sigma}\bar{X}-\frac{\mu\sqrt{n}}{\sigma}\\
\\
\Rightarrow M_{Y}\left(t\right)&=e^{-\frac{\mu\sqrt{n}}{\sigma}t}M_{\bar{X}}\left ( \frac{\sqrt{n}}{\sigma}t\right)=e^{-\frac{\mu\sqrt{n}}{\sigma}t}\left [M_{X}\left ( \frac{\sqrt{n}}{n\sigma}t\right)\right]^{n}=e^{-\frac{\mu\sqrt{n}}{\sigma}t}\left [M_{X}\left ( \frac{t}{\sqrt{n}\sigma}\right)\right]^{n}\\
\\
\Rightarrow ln\left [M_{Y}\left(t\right)\right]&=-\frac{\mu\sqrt{n}}{\sigma}t+n\cdot ln\left [M_{X}\left ( \frac{t}{\sqrt{n}\sigma}\right)\right] (令 p=\frac{t}{\sqrt{n}\sigma})\\
\\
&=-\frac{\mu t^{2}}{p\sigma^{2}}+\frac{t^{2}}{p^{2}\sigma^{2}}ln\left [M_{X}\left(p\right)\right]=\frac{t^{2}}{\sigma^{2}}\left ( \frac{ln\left [M_{X}\left(p\right)\right]-\mu p}{p^{2}} \right )\\
\\
\Rightarrow \lim_{n\rightarrow \infty}ln\left [M_{Y}\left(t\right)\right]&=\lim_{p\rightarrow 0}\frac{t^{2}}{\sigma^{2}}\left ( \frac{ln\left [M_{X}\left(p\right)\right]-\mu p}{p^{2}} \right )\\
\\
&=\lim_{p\rightarrow 0}\frac{t^{2}}{\sigma^{2}}\left ( \frac{\frac{1}{M_{X}\left(p\right)}\cdot \frac{M_{X}\left(p\right)}{dp}-\mu}{2p} \right )=\lim_{p\rightarrow 0}\frac{t^{2}}{\sigma^{2}}\left ( \frac{\frac{M^{'}_{X}\left(p\right)}{M_{X}\left(p\right)}-\mu}{2p} \right )\\
\\
&=\lim_{p\rightarrow 0}\frac{t^{2}}{\sigma^{2}}\left ( \frac{\frac{M^{''}_{X}\left(p\right)\cdot M_{X}\left(p\right)-M^{'}_{X}\left(p\right)\cdot M^{'}_{X}\left(p\right)}{\left ( M_{X}\left(p\right) \right )^{2}}}{2}\right)=\lim_{p\rightarrow 0}\frac{t^{2}}{\sigma^{2}}\left ( \frac{\mu^{'}_{2}-\mu^{2}}{2}\right)\\
\\
&=\frac{t^{2}}{\sigma^{2}}\left ( \frac{\sigma^{2}}{2}\right)=\frac{t^{2}}{2}
\end{aligned}
\end{equation}$$

$\therefore \ n \rightarrow \infty,\ M_{Y}\left ( t \right )\rightarrow e^{\frac{t^{2}}{2}}$ (標準常態分配的動差生成函數)

$\Rightarrow Y=\frac{\bar{X}-\mu}{\sigma/\sqrt{n}}\sim N\left ( 0,1 \right )$

$$\therefore \lim_{n\rightarrow \infty }\frac{X_{1}+X_{2}+...+X_{n}}{n}\sim N\left ( \mu ,\frac{\sigma }{\sqrt{n}} \right )$$

#### 六. T 分配

(1) T 分配的來源:

當大樣本($n \geq 30$)的 $\sigma$ 未知時，可以用 $s$ 表示，此時統計量 $\frac{\bar{X}-\mu}{\sigma/\sqrt{n}}$ 改成 $\frac{\bar{X}-\mu}{s/\sqrt{n}}$ (T 分配)會接近**標準常態分配(中央極限定理)**。若 $n<30$，則不再服從標準常態分配。

(2) 定義: **(標準常態分配與卡方分配之組合)**

$$T=\frac{\bar{X}-\mu}{s/\sqrt{n}}=\frac{\left ( \bar{X}-\mu \right )/\left ( \sigma /\sqrt{n} \right )}{\left ( s/\sqrt{n} \right )/\left ( \sigma /\sqrt{n} \right )}=\frac{Z}{\sqrt{s^{2}/\sigma^{2}}}=\frac{Z}{\sqrt{\chi_{n-1}^{2}/\left ( n-1 \right )}}$$

$$(Note: \chi_{n-1}^{2}\sim \frac{\left ( n-1 \right )s^{2}}{\sigma^{2}})$$

(3) T 分配的機率密度函數: 

$$f\left ( t \right )=\frac{\Gamma \left ( \frac{\nu+1}{2} \right )}{\sqrt{\pi \nu}\Gamma\left (\frac{\nu}{2} \right )}\left ( 1+\frac{t^{2}}{\nu} \right )^{-\left ( \frac{\nu+1}{2} \right )},\ -\infty <t<\infty $$

其中自由度 ν 為正整數。

**推導步驟:**

i. 寫出兩個分配的聯合機率密度函數

ii. 使用變數變換與 Jacobian

iii. 利用邊際機率密度函數求出 T 分配的機率密度函數

$$f_{1}\left ( z \right )=\frac{1}{\sqrt{2\pi }}e^{-\frac{z^{2}}{2}};\ f_{2}\left ( x \right )=\frac{1}{2^{\nu/2}\Gamma \left ( \nu/2 \right )}x^{\frac{\nu}{2}-1}e^{-\frac{x}{2}}$$

$$\Rightarrow g\left ( z,x \right )=\frac{1}{\sqrt{2\pi }}e^{-\frac{z^{2}}{2}}\cdot \frac{1}{2^{\nu/2}\Gamma \left ( \nu/2 \right )}x^{\frac{\nu}{2}-1}e^{-\frac{x}{2}};\ \ -\infty <z<\infty,\ 0<x<\infty$$

$$令\ t=\frac{z}{\sqrt{x/\nu}},\ y=x \Rightarrow z=\frac{t\sqrt{y}}{\sqrt{\nu}}$$

$$J=\frac{\partial \left ( z,x \right )}{\partial \left ( t,y \right )}=\begin{vmatrix}
\frac{\partial z}{\partial t} & \frac{\partial z}{\partial y}\\ 
\frac{\partial x}{\partial t} & \frac{\partial x}{\partial y}
\end{vmatrix}
=\begin{vmatrix}
\frac{\sqrt{y}}{\sqrt{\nu}} & \frac{t}{2\sqrt{y}\sqrt{\nu}}\\ 
0 & 1
\end{vmatrix}
=\frac{\sqrt{y}}{\sqrt{\nu}}$$

$$\begin{equation} 
\begin{aligned}
\Rightarrow h\left ( t,y \right )&=\frac{1}{\sqrt{2\pi }}e^{-\frac{t^{2}y}{2\nu }}\cdot \frac{1}{2^{\nu/2}\Gamma \left ( \nu/2 \right )}y^{\frac{\nu}{2}-1}e^{-\frac{y}{2}}\cdot \frac{\sqrt{y}}{\sqrt{\nu}};\ -\infty <t< \infty,\ 0<y<\infty\\
&=\frac{1}{\sqrt{\nu}\sqrt{2\pi }}\frac{1}{2^{\nu/2}\Gamma \left ( \nu/2 \right )}y^{\frac{\nu+1}{2}-1}e^{-\frac{y}{2}\left ( 1+\frac{t^{2}}{\nu} \right )}\\
\\
&令\ w=\frac{y}{2}\left ( 1+\frac{t^{2}}{\nu} \right ) \Rightarrow dy=\left ( \frac{2\nu}{\nu+t^{2}} \right )dw\\
\\
\Rightarrow h\left ( t \right )&=\int_{0}^{\infty}\frac{1}{\sqrt{\nu}\sqrt{2\pi }}\frac{1}{2^{\nu/2}\Gamma \left ( \nu/2 \right )}y^{\frac{\nu+1}{2}-1}e^{-\frac{y}{2}\left ( 1+\frac{t^{2}}{\nu} \right )}dy\\
&=\int_{0}^{\infty}\frac{1}{\sqrt{\nu}\sqrt{2\pi }}\frac{1}{2^{\nu/2}\Gamma \left ( \nu/2 \right )}\left ( \frac{2\nu w}{\nu+t^{2}} \right )^{\frac{\nu+1}{2}-1}\cdot e^{-w}\left ( \frac{2\nu}{\nu+t^{2}} \right )dw\\
&=\frac{1}{\sqrt{\nu}\sqrt{2\pi }}\frac{1}{2^{\nu/2}\Gamma \left ( \nu/2 \right )}\left ( \frac{2\nu}{\nu+t^{2}} \right )^{\frac{\nu+1}{2}}\int_{0}^{\infty}w^{\frac{\nu+1}{2}-1}\cdot e^{-w}dw\\
&=\frac{1}{\sqrt{\nu}\sqrt{2\pi }}\frac{1}{2^{\nu/2}\Gamma \left ( \nu/2 \right )}\left ( \frac{2\nu}{\nu+t^{2}} \right )^{\frac{\nu+1}{2}}\Gamma \left ( \frac{\nu+1}{2} \right )\\
&=\frac{1}{\sqrt{\nu \pi}\Gamma \left ( \nu/2 \right )}\left ( \frac{\nu}{\nu+t^{2}} \right )^{\frac{\nu+1}{2}}\Gamma \left ( \frac{\nu+1}{2} \right )\\
&=\frac{\Gamma \left ( \frac{\nu+1}{2} \right )}{\sqrt{\pi \nu}\Gamma\left (\frac{\nu}{2} \right )}\left ( 1+\frac{t^{2}}{\nu} \right )^{-\left ( \frac{\nu+1}{2} \right )}
\end{aligned}
\end{equation}$$

#### 七. F 分配

(1) 定義: 兩個卡方隨機變數各自除以自由度之比值

$$ F=\frac{\chi_{1}^{2}/\nu_{1}}{\chi_{2}^{2}/\nu_{2}} $$

(2) F 分配的機率密度函數: 

$$ h\left ( f \right )=\frac{\Gamma \left ( \frac{\nu_{1}+\nu_{2}}{2} \right )\cdot \left ( \frac{\nu_{1}}{\nu_{2}} \right )^{\frac{\nu_{1}}{2}}}{\Gamma \left ( \frac{\nu_{1}}{2} \right )\Gamma \left ( \frac{\nu_{2}}{2} \right )}\cdot \frac{f^{\frac{\nu_{1}}{2}-1}}{\left ( 1+\frac{\nu_{1}}{\nu_{2}}f \right )^{\left ( \nu_{1}+\nu_{2} \right )/2}} $$

**a. 推導:**

考慮兩個獨立卡方分配 $f_{1}\left ( s \right )$、$f_{2}\left ( t \right )$ 之聯合機率密度函數 $g\left ( s,t \right )$:

$$f_{1}\left ( s \right )=\frac{1}{2^{\frac{\nu_{1}}{2}}\Gamma \left ( \frac{\nu_{1}}{2} \right )}s^{\frac{\nu_{1}}{2}-1}e^{-\frac{s}{2}};\ f_{2}\left ( t \right )=\frac{1}{2^{\frac{\nu_{2}}{2}}\Gamma \left ( \frac{\nu_{2}}{2} \right )}t^{\frac{\nu_{2}}{2}-1}e^{-\frac{t}{2}}$$

$$\begin{equation} 
\begin{aligned}
\Rightarrow g\left ( s,t \right )&=f_{1}\left ( s \right )\cdot f_{2}\left ( t \right )=\frac{1}{2^{\frac{\nu_{1}}{2}}\Gamma \left ( \frac{\nu_{1}}{2} \right )}s^{\frac{\nu_{1}}{2}-1}e^{-\frac{s}{2}}\cdot \frac{1}{2^{\frac{\nu_{2}}{2}}\Gamma \left ( \frac{\nu_{2}}{2} \right )}t^{\frac{\nu_{2}}{2}-1}e^{-\frac{t}{2}}\\
&=\frac{1}{2^{\frac{\nu_{1}+\nu_{2}}{2}}\Gamma \left ( \frac{\nu_{1}}{2} \right )\Gamma \left ( \frac{\nu_{2}}{2} \right)}s^{\frac{\nu_{1}}{2}-1}\cdot t^{\frac{\nu_{2}}{2}-1}\cdot e^{-\frac{s+t}{2}}\\
\\
&使用變數變換\left\{\begin{matrix}
f=&\frac{s/\nu_{1}}{t/\nu_{2}}\\ 
w=&t
\end{matrix}\right.
\Rightarrow
\left\{\begin{matrix}
s=&\left ( \frac{\nu_{1}}{\nu_{2}} \right )\cdot wf\\ 
t=&w
\end{matrix}\right.\\
\\
J&=\frac{\partial \left ( s,t \right )}{\partial \left ( f,w \right )}=\begin{vmatrix}
\frac{\partial s}{\partial f} & \frac{\partial s}{\partial w}\\ 
\frac{\partial t}{\partial f} & \frac{\partial t}{\partial w}
\end{vmatrix}
=\begin{vmatrix}
\left ( \frac{\nu_{1}}{\nu_{2}} \right )w & \left ( \frac{\nu_{1}}{\nu_{2}} \right )f\\ 
0 & 1
\end{vmatrix}
=\left ( \frac{\nu_{1}}{\nu_{2}} \right )w\\
\\
\Rightarrow h\left ( f,w \right )&=\frac{1}{2^{\frac{\nu_{1}+\nu_{2}}{2}}\Gamma \left ( \frac{\nu_{1}}{2} \right )\Gamma \left ( \frac{\nu_{2}}{2} \right)}\left ( \frac{\nu_{1}fw}{\nu_{2}} \right )^{\frac{\nu_{1}}{2}-1}\cdot w^{\frac{\nu_{2}}{2}-1}\cdot e^{-\frac{1}{2}\left ( \frac{\nu_{1}fw}{\nu_{2}} +w\right )}\cdot \left ( \frac{\nu_{1}w}{\nu_{2}} \right )\\
&=\frac{1}{2^{\frac{\nu_{1}+\nu_{2}}{2}}\Gamma \left ( \frac{\nu_{1}}{2} \right )\Gamma \left ( \frac{\nu_{2}}{2} \right)}\left ( \frac{\nu_{1}w}{\nu_{2}} \right )^{\frac{\nu_{1}}{2}}\cdot f^{\frac{\nu_{1}}{2}-1}\cdot w^{\frac{\nu_{2}}{2}-1}\cdot e^{-\frac{1}{2}\left ( \frac{\nu_{1}fw}{\nu_{2}} +w\right )};\ 0<f,w<\infty\\
\\
\Rightarrow h\left ( f \right )&=\frac{1}{2^{\frac{\nu_{1}+\nu_{2}}{2}}\Gamma \left ( \frac{\nu_{1}}{2} \right )\Gamma \left ( \frac{\nu_{2}}{2} \right)}\cdot f^{\frac{\nu_{1}}{2}-1}\cdot \int_{0}^{\infty}\left ( \frac{\nu_{1}w}{\nu_{2}} \right )^{\frac{\nu_{1}}{2}}\cdot w^{\frac{\nu_{2}}{2}-1}\cdot e^{-\frac{1}{2}\left ( \frac{\nu_{1}fw}{\nu_{2}} +w\right )} dw\\
&=\frac{1}{2^{\frac{\nu_{1}+\nu_{2}}{2}}\Gamma \left ( \frac{\nu_{1}}{2} \right )\Gamma \left ( \frac{\nu_{2}}{2} \right)}\cdot f^{\frac{\nu_{1}}{2}-1}\cdot \left ( \frac{\nu_{1}}{\nu_{2}} \right )^{\frac{\nu_{1}}{2}}\cdot  \int_{0}^{\infty} w^{\frac{\nu_{1}+\nu_{2}}{2}-1}\cdot e^{-\frac{1}{2}\left ( \frac{\nu_{1}f}{\nu_{2}} +1\right )w} dw\\
\\
&令\ y=\frac{1}{2}\left ( \frac{\nu_{1}f}{\nu_{2}} +1\right )w \Rightarrow dw=\frac{2}{\frac{\nu_{1}f}{\nu_{2}}+1}dy\\
&=\frac{1}{2^{\frac{\nu_{1}+\nu_{2}}{2}}\Gamma \left ( \frac{\nu_{1}}{2} \right )\Gamma \left ( \frac{\nu_{2}}{2} \right)}\cdot f^{\frac{\nu_{1}}{2}-1}\cdot \left ( \frac{\nu_{1}}{\nu_{2}} \right )^{\frac{\nu_{1}}{2}}\cdot  \int_{0}^{\infty} \left ( \frac{2}{\frac{\nu_{1}f}{\nu_{2}}+1} \right )^{\frac{\nu_{1}+\nu_{2}}{2}-1}\cdot y^{\frac{\nu_{1}+\nu_{2}}{2}-1}\cdot e^{-y}\cdot \frac{2}{\frac{\nu_{1}f}{\nu_{2}}+1}dy\\
&=\frac{1}{2^{\frac{\nu_{1}+\nu_{2}}{2}}\Gamma \left ( \frac{\nu_{1}}{2} \right )\Gamma \left ( \frac{\nu_{2}}{2} \right)}\cdot f^{\frac{\nu_{1}}{2}-1}\cdot \left ( \frac{\nu_{1}}{\nu_{2}} \right )^{\frac{\nu_{1}}{2}}\cdot \left ( \frac{2}{\frac{\nu_{1}f}{\nu_{2}}+1} \right )^{\frac{\nu_{1}+\nu_{2}}{2}}\cdot \int_{0}^{\infty} y^{\frac{\nu_{1}+\nu_{2}}{2}-1}\cdot e^{-y}dy\\
&=\frac{1}{\Gamma \left ( \frac{\nu_{1}}{2} \right )\Gamma \left ( \frac{\nu_{2}}{2} \right)}\cdot f^{\frac{\nu_{1}}{2}-1}\cdot \left ( \frac{\nu_{1}}{\nu_{2}} \right )^{\frac{\nu_{1}}{2}}\cdot \left ( \frac{1}{\frac{\nu_{1}f}{\nu_{2}}+1} \right )^{\frac{\nu_{1}+\nu_{2}}{2}}\cdot \Gamma \left ( \frac{\nu_{1}+\nu_{2}}{2} \right )\\
&=\frac{\Gamma \left ( \frac{\nu_{1}+\nu_{2}}{2} \right )\cdot \left ( \frac{\nu_{1}}{\nu_{2}} \right )^{\frac{\nu_{1}}{2}}}{\Gamma \left ( \frac{\nu_{1}}{2} \right )\Gamma \left ( \frac{\nu_{2}}{2} \right )}\cdot \frac{f^{\frac{\nu_{1}}{2}-1}}{\left ( 1+\frac{\nu_{1}}{\nu_{2}}f \right )^{\left ( \nu_{1}+\nu_{2} \right )/2}}
\end{aligned}
\end{equation}$$

i. 右偏

ii. 兩個自由度

(3) F 分配的性質:

a. $f\left ( \nu_{1},\nu_{2} \right ) = \frac{1}{f\left ( \nu_{2},\nu_{1} \right )}$

說明:

$$f\left ( \nu_{1},\nu_{2} \right )=\frac{\chi_{1}^{2}/\nu_{1}}{\chi_{2}^{2}/\nu_{2}}=\frac{1}{\left ( \frac{\chi_{2}^{2}}{\nu_{2}} \right )/\left ( \frac{\chi_{1}^{2}}{\nu_{1}} \right )}=\frac{1}{f\left ( \nu_{2},\nu_{1} \right )}$$

b. $f_{1-\alpha}\left ( \nu_{2},\nu_{1} \right )=\frac{1}{f_{\alpha}\left ( \nu_{1},\nu_{2} \right )}$

說明:

$$\begin{equation} 
\begin{aligned}
& P\left [ f\left ( \nu_{1},\nu_{2} \right )>f_{\alpha }\left ( \nu_{1},\nu_{2} \right ) \right ]=\alpha \\
\\
\Rightarrow\ & P\left [ f\left ( \nu_{1},\nu_{2} \right )<f_{\alpha }\left ( \nu_{1},\nu_{2} \right ) \right ]=1-\alpha\\
\\
\Rightarrow\ & P\left [ \frac{1}{f\left ( \nu_{2},\nu_{1} \right )}<f_{\alpha }\left ( \nu_{1},\nu_{2} \right ) \right ]=1-\alpha\\
\\
\Rightarrow\ & P\left [ f\left ( \nu_{2},\nu_{1} \right )>\frac{1}{f_{\alpha }\left ( \nu_{1},\nu_{2} \right )} \right ]=1-\alpha,\ (i)\\
\\
&P\left [ f\left ( \nu_{1},\nu_{2} \right )<f_{1-\alpha }\left ( \nu_{1},\nu_{2} \right ) \right ]=1-\alpha\\
\\
\Rightarrow\ & P\left [ \frac{1}{f\left ( \nu_{2},\nu_{1} \right )}<\frac{1}{f_{1-\alpha }\left ( \nu_{2},\nu_{1} \right )} \right ]=1-\alpha\\
\\
\Rightarrow\ & P\left [ f\left ( \nu_{2},\nu_{1} \right )>f_{1-\alpha }\left ( \nu_{2},\nu_{1} \right ) \right ]=1-\alpha,\ (ii)\\
\\
&(i)=(ii) \Rightarrow f_{1-\alpha}\left ( \nu_{2},\nu_{1} \right )=\frac{1}{f_{\alpha}\left ( \nu_{1},\nu_{2} \right )}
\end{aligned}
\end{equation}$$

c. $F\left ( 1,\nu_{2} \right )\sim T^{2}\left ( \nu_{2} \right )$

說明:

$$T=\frac{Z}{\sqrt{\frac{\chi_{\nu_{2}}^{2}}{\nu_{2}}}}
\Rightarrow T^{2}=\frac{Z^{2}}{\frac{\chi_{\nu_{2}}^{2}}{\nu_{2}}}=\frac{\frac{\chi_{1}^{2}}{1}}{\frac{\chi_{\nu_{2}}^{2}}{\nu_{2}}}\sim F\left ( 1,\nu_{2} \right )$$

(4) F 分配的期望值:

a. 已知卡方分配的期望值 $E\left ( x \right )=\nu$，則 $E\left ( \frac{1}{x} \right )=\frac{1}{\nu-2}$

**推導:**

$$\begin{equation} 
\begin{aligned}
E\left ( \frac{1}{x} \right )&=\int_{0}^{\infty}\frac{1}{x}\cdot \frac{1}{2^{\frac{\nu}{2}}\Gamma \left ( \frac{\nu}{2} \right )}\cdot x^{\frac{\nu}{2}-1}\cdot e^{-\frac{x}{2}}dx\\
\\
&(令\ y=\frac{x}{2} \Rightarrow dx=2dy)\\
\\
&=\int_{0}^{\infty}\frac{1}{2y}\cdot \frac{1}{2^{\frac{\nu}{2}}\Gamma \left ( \frac{\nu}{2} \right )}\cdot \left ( 2y \right )^{\frac{\nu}{2}-1}\cdot e^{-y}\cdot 2dy\\
\\
&=\frac{1}{2\cdot \Gamma \left ( \frac{\nu}{2} \right )}\cdot \int_{0}^{\infty}y^{\left ( \frac{\nu}{2}-1 \right )-1}\cdot e^{-y}dy\\
\\
&=\frac{1}{2\cdot \Gamma \left ( \frac{\nu}{2} \right )}\cdot \Gamma \left ( \frac{\nu}{2}-1 \right )\\
\\
&=\frac{1}{2\cdot \left ( \frac{\nu}{2}-1 \right )\cdot \Gamma \left ( \frac{\nu}{2}-1 \right )}\cdot \Gamma \left ( \frac{\nu}{2}-1\right )\\
\\
&=\frac{1}{\nu-2}
\end{aligned}
\end{equation}$$


b. 計算 F 分配的期望值 $E(F)$:

$$E\left ( F \right )=E\left ( \frac{\chi^{2}_{1}/\nu_{1}}{\chi^{2}_{2}/\nu_{2}} \right )=\frac{\nu_{2}}{\nu_{1}}\cdot E\left ( \chi_{1}^{2} \right )\cdot E\left ( \frac{1}{\chi_{2}^{2}} \right )=\frac{\nu_{2}}{\nu_{1}}\cdot \nu_{1}\cdot \frac{1}{\nu_{2}-2}=\frac{\nu_{2}}{\nu_{2}-2}$$
